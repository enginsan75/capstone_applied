```python
filepath = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
df = pd.read_csv(filepath, header=None)
```

```python
# create headers list
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

```python
df.columns = headers
df.columns
```

```python
df.head(10)
```

```python
df1=df.replace('?',np.NaN) #replace ? with NaN so dropna() can work
```

```python
df1.head(10)
```

```python
df2=df1.dropna(subset=["price"], axis=0).copy() #all rows with na will be dropped in price column
df2.head(20)
```

```python
df2[['price']] = df2[['price']].apply(pd.to_numeric)
```

```python
df2['price'].dtype
```

```python
df2[['normalized-losses']] = df2[['normalized-losses']].apply(pd.to_numeric)
```

```python
mean_of_normalizedloss = df2['normalized-losses'].mean()
mean_of_normalizedloss
```

```python
#replace NANs with mean:
df2['normalized-losses'].replace(np.nan, mean_of_normalizedloss, inplace = True)
```

```python
df2['normalized-losses']
```

```python
#rename the column:
df2.rename(columns={"num-of-cylinders": "number-of-cylinders"},inplace = True) 
                    
```

```python
df2["number-of-cylinders"]
```

```python
#change dt type of column integer to float:
df2["curb-weight"]=df2["curb-weight"].astype("float")
```

```python
df2["curb-weight"].dtype
```

```python
df2["curb-weight"].round(decimals=3)
```

```python
#DATA NORMALIZATION:
#1.Simple feature scaling by dividing max value:
df2["height"] = df2["height"]/df2["height"].max()
```

```python
#2.Scaling by min-max method:

df2["length"] = (df2["length"]-df2["length"].min())/(df2["length"].max()-df2["length"].min())
```

```python
# Z-score method:
df2["width"] = (df2["width"]-df2["width"].mean())/(df2["width"].std())
```

```python
#BINNING: GROUPING INTO RANGES OR BINS
#Price column into Low, Medium, High bins:

bins = np.linspace(min(df2["price"]), max(df2["price"]), 4) #4 means equal distance
```
